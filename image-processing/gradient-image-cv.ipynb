# Morphological: Gradients

######  An image gradient is a direction change in the intensity or color in an image


###### For more details check out the wikipedia article on the Sobel Operator for full math details: https://en.wikipedia.org/wiki/Sobel_operator



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Show Image Function
def display_img(img):
    fig = plt.figure(figsize=(12, 12))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
# load the sudoku image and gray scale
img = cv2.imread('/Users/neemiasbsilva/Downloads/Computer-Vision-with-Python/DATA/sudoku.jpg', 0)
```

```python
display_img(img)
```

```python
# calculating x-sobel
# Depth is the "precision" of each pixel. Typically it can be 8/24/32 bit for displaying
# dx = 1
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx)
```

```python
# calculating x-sobel
# Depth is the "precision" of each pixel. Typically it can be 8/24/32 bit for displaying
# dx = 0
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

```python
display_img(sobely)
```
