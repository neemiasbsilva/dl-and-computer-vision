# Using Mathematical Morpholoy (MM) to Analysis and Processing of Geometical Structures

```python
# import some libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load The Image
def load_img():
    blank_img = np.zeros((600, 600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text='ABCDE', org=(50, 300), fontFace=font, fontScale= 5, color=(255, 255, 255), thickness=20)
    return blank_img
```

```python
# Show The Image
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

**Applied Erosion Operator**

```python
kernel = np.ones((5, 5), dtype=np.uint8)
kernel
```

```python
result = cv2.erode(img, kernel, iterations=2)
```

```python
display_img(result)
```

**Applied to Open Operator**

```python
img = load_img()
```

```python
white_noise = np.random.randint(low=0, high=2, size=(600, 600))
```

```python
white_noise
```

```python
display_img(white_noise)
```

```python
# putting this noise in original image
img.max()
```

```python
white_noise = white_noise * 255
```

```python
display_img(white_noise)
```

```python
noise_img = white_noise + img
```

```python
display_img(noise_img)
```

```python
# Using dilation

opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)

display_img(opening)
```

**Applied Close Operation**

```python
img = load_img()
```

```python
black_noise = np.random.randint(low=0, high=2, size=(600, 600))
```

```python
black_noise *= -255
```

```python
black_noise
```

```python
black_noise = img + black_noise
```

```python
black_noise[black_noise==-255] = 0
display_img(black_noise)
```

```python
clossing = cv2.morphologyEx(black_noise, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(clossing)
```

**Morphological Gradient**

```python
img = load_img()
```

```python
display_img(img)
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
```

```python
display_img(gradient)
```
